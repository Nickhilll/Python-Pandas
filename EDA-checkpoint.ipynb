## Types of Data

## 1.univariate analysis

### Analyze one column feature at a time

```python
cat
```

```python
for i in cat:
    plt.figure(figsize = (10, 5))
    plt.title(f'countplot for {i}')
    sns.countplot(data = df, x = i)
    plt.show()
```

## Coutinuous -  histogram

#### kde - kernel density estimate

```python
for i in con:
    plt.figure(figsize = (10,5))
    plt.title(f'Histograme for {i}')
    sns.histplot(data = df, x = i, kde = True)
    plt.show()
```
